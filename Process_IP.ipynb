```python
ips = []
for file in ip_files:
    f = open(path+'/ips_random/'+file, "r")
    for line in f.readlines():
        ip = re.findall( r'[0-9]+(?:\.[0-9]+){3}', line )
        ips.append(ip)
    f.close()
```
